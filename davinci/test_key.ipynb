```python
def extract_last_number(string):
    pattern = r'\d+'  # 匹配一个或多个数字
    matches = re.findall(pattern, string)
    if matches:
        return matches[-1]  # 返回最后一个匹配到的数字
    else:
        return None  # 如果没有匹配到数字，则返回None
```
